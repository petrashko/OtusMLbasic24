# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = np.zeros(10)
arr[4] = 1
arr = arr.reshape(2, 5)
arr
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10, 50)
arr = arr[::-1]
print(arr)
mask = (arr % 2) == 0
arr2 = arr[mask]
arr2
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr = np.arange(0, 9).reshape(3, 3)
arr
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.rand(4, 3, 2)
print(arr)
min_val = arr.flatten().min()
max_val = arr.flatten().max()
min_val, max_val
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr1 = np.random.randint(1, 11, (6, 4))
arr2 = np.random.randint(1, 11, (4, 3))
print(arr1)
print(arr2)
np.dot(arr1, arr2)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.randint(0, 101, (7, 7))
print(arr)
mean_val = arr.mean()
std_val = arr.std()
print(f'mean = {mean_val}; std = {std_val}')
arr_norm = (arr - mean_val) / std_val
arr_norm
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
group_by_day = df.groupby('day').mean(numeric_only=True)
group_by_day['total_bill']
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
mask = df['total_bill'] > df['total_bill'].median()
df1 = df[mask]
group_by_sex = df1.groupby('sex').mean(numeric_only=True)
group_by_sex['tip']
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 4), dpi=80)
sns.histplot(data=df, x='total_bill');
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 4), dpi=80)
sns.scatterplot(data=df, x='total_bill', y='tip', alpha=0.9)
axes.set_xlabel('Общая сумма счета')
axes.set_ylabel('Размер чаевых');
```

### Постройте pairplot

```python
sns.pairplot(
    data=df,
    corner=True,
    hue='smoker'
);
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 6), dpi=80)
sns.boxplot(data=df, y='total_bill', x='day')
axes.set_xlabel('День недели')
axes.set_ylabel('Общая сумма счета')
axes.grid(axis='y');
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 4), dpi=80)
sns.histplot(data=df, x='tip', hue='time');
```
